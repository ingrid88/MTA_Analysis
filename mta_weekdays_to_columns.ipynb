```python
import numpy as np
tmp = pd.get_dummies(geo_pd[['STATION_x', 'DATE', 'COUNT', 'LAT','LON']], columns = [ 'DATE'])
for col in tmp.columns:
    if col[:4] == 'DATE':
        tmp[col] = np.array(tmp[col])*np.array(map(int,tmp['COUNT']))
```

```python
sublist = tmp.drop_duplicates(subset="STATION_x").iloc[:,0:4]
sublist = sublist.drop('COUNT', 1)

# df1.iloc[:,1:3]
```
