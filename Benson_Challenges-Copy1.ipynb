```python
def mta_dict(df):
    # retrieve first four columns as tuples
    keys = zip(df["C/A"],df["UNIT"],df["SCP"],df["STATION"])
    
    # retrieve all other columns as a panda dataframe
    value_columns = df.iloc[:,4:]
    
    # add tuple keys to value_columns df
    value_columns["KEY"] = keys
    
    # group by tuple Key
    grouped_df = value_columns.groupby("KEY")
    
    dictionary = {}
    for name,group in grouped_df:
        sub = map(list, group.iloc[:,:7].values)
        dictionary[name] = sub
    return dictionary

dictionary = mta_dict(df)
```

```python
def mta_datetime_dict(df):
    
    # turn date and time strings into datetime objects
    datetimes = pd.to_datetime(df["DATE"]+df["TIME"],format="%m/%d/%Y%H:%M:%S")
    
    # build key tuples list
    keys = zip(df["C/A"],df["UNIT"],df["SCP"],df["STATION"])
    
    # turn entries series into a dataframe 
    entries = df['ENTRIES'].to_frame()
    
    #append the keys 
    entries["keys"]=keys
    entries["datetimes"]=datetimes
    grouped = entries.groupby("keys")
    
    # create dictionary by building a list out of group elements as values
    dictionary = {}
    for name, group in grouped:
        sub = map(list, group.loc[:,["datetimes","ENTRIES"]].values)
        dictionary[name] = sub
    return dictionary


## PIPELINE
datetime_dict = mta_datetime_dict(df) 
```
